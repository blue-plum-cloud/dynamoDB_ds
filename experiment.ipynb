```python
#### Multiple clients write something x10 and read it (concurrently).
### Various Num of clients and Number of Nodes (N fixed 10)

import matplotlib.pyplot as plt
import numpy as np

# Your data
x_axis = ['20', '40', '60', '80', '100', '120']
y_1 = ['57.805208', '88.126708', '232.957125', '279.996792', '421.013417', '324.102791']
y_2 = ['65.060417','116.244292','224.585541','318.9765','398.597542','465.600792']
y_3 = ['80.019916', '132.695125','346.845042','342.698416','420.807625','497.164666']


# Convert data to numerical format
x = np.arange(len(x_axis))  # the label locations
y_1 = [float(i) for i in y_1]
y_2 = [float(i) for i in y_2]
y_3 = [float(i) for i in y_3]

# Bar width
width = 0.25  # the width of the bars

# Plotting
fig, ax = plt.subplots()
rects1 = ax.bar(x - width, y_1, width, label='20 Nodes')
rects2 = ax.bar(x, y_2, width, label='40 Nodes')
rects3 = ax.bar(x + width, y_3, width, label='60 Nodes')

# Add some text for labels, title, and custom x-axis tick labels, etc.
ax.set_xlabel('Number of Clients')
plt.ylabel('Time Taken (ms)')
plt.title('Multiple Clients Read and Write Concurrently')
plt.legend(title='NumOfClients: 20-120\nNumOfNodes: 20,40,60\nNumOfTokens: 20\nN: 10R: 10\nW: 10', loc='upper left')
ax.set_xticks(x)
ax.set_xticklabels(x_axis)
# ax.legend()

plt.show()
```
